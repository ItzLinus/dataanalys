```python
# Importerar libraries - pandas, matplotlib, numpy
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

# Funktioner
def importAllCases():                   # Läser in dokument, extraherar information och samlar alla tal
    dfCases = pd.read_csv("Regional_Daily_Cases.csv")
    totCases = dfCases['Sweden_Total_Daily_Cases'].sum()

    return totCases

def importDeaths():                     # Läser in dokument, extraherar information och samlar alla tal
    dfDeaths = pd.read_csv("National_Daily_Deaths.csv")
    totalDeaths = dfDeaths['National_Daily_Deaths'].sum()
    
    return totalDeaths


def Diagrammer(a, b):                   # Funktion för att skapa ett diagram.  
    labels = 'Death', 'Alive' 
    sizes = [a, b]
    explode = (0.09, 0)

    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, 
            explode=explode, 
            labels=labels, 
            autopct='%1.1f%%',
            shadow=True, 
            startangle=90)              # Konfigurationen för hur diagrammer kommer att ser ut 
    ax1.axis('equal')
    
    plt.title("Deaths / Alive ratio ")               #titel för pie charten
    plt.show()  




Diagrammer(importDeaths(), importAllCases()) # Här kör jag de tre funktionerna tillsammans, jag använder "Diagrammer(a, b)" sedan använder jag importDeaths och ImportAllCases för att skapa ett diagram med hjälp av datan. 
```

```python
# Importerar libraries - plotly_express, pandas. Även libraries för att kunna bilda grafer.  
import plotly_express as px
import pandas as pd 
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Jag importerar även Dash för att skapa min dashboard 
import dash
import dash_core_components as dcc
import dash_html_components as html



dfNational = pd.read_csv('National_Daily_Deaths.csv') 
dfGender = pd.read_csv("Gender_Data.csv")
dfTotalDeaths = pd.read_csv("National_Total_Deaths_by_Age_Group.csv")

# Skapar variabler för kön, alla fall, ICU fall och dödssiffor.
Graph1_gender = dfGender['Gender']
Graph1_totalcase = dfGender['Total_Cases']
Graph1_totalicu = dfGender['Total_ICU_Admissions']
Graph1_totaldeath = dfGender['Total_Deaths']

Graph2_Date = dfNational['Date']
Graph2_Daily = dfNational['National_Daily_Deaths']

Graph3_Age = dfTotalDeaths['Age_Group']
Graph3_TotalCase = dfTotalDeaths['Total_Cases']
Graph3_Deaths = dfTotalDeaths['Total_Deaths']

# Här skapar jag subplots dvs "små grafer". Jag lägger även till titlar till varje graf. 
fig = make_subplots(
    rows=3, cols=3,
    specs=[[{}, {}, {}],
           [{"colspan":3}, None, None],
           [{"colspan":3}, None, None]],

        print_grid=False,
    subplot_titles=("Total Cases","Total Death","Total ICU", "Deaths per day", "Deaths per age"))

# Nedan är vad som behövs för att skapa varje diagram. Även hur layouten på alla diagram ska se ut. 
fig.add_trace(go.Bar(
    x=Graph1_gender, 
    y=Graph1_totalcase),
    row=1, col=1)
fig.add_trace(go.Bar(
    x=Graph1_gender, 
    y=Graph1_totaldeath), 
    row=1, col=2)
fig.add_trace(go.Scatter(
    x=Graph1_gender, 
    y=Graph1_totalicu), 
    row=1, col=3)
fig.add_trace(go.Box(
    x=Graph2_Date, 
    y=Graph2_Daily), 
    row=2, col=1)
fig.add_trace(go.Line(
    x=Graph3_Age, 
    y=Graph3_TotalCase),
    row=3, col=1)

fig.update_layout(
    showlegend=False,
    title_text="Data Analys Covid-19",
    height=700)

# Titlar för graferna
fig.update_xaxes(title_text="Kön", row=1, col=1)
fig.update_xaxes(title_text="Kön", row=1, col=2)
fig.update_xaxes(title_text="Kön", row=1, col=3)
fig.update_xaxes(title_text="Datum", row=2, col=1)
fig.update_xaxes(title_text="Ålder", row=3, col=1)


# Visar grafen 
# fig.show()

# Funktioner för dash
app = dash.Dash()
app.layout = html.Div([dcc.Graph(figure=fig)])
  
app.run_server(debug=True, use_reloader=False) 
```
