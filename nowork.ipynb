```python
# Importerar libraries - pandas, matplotlib, numpy
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

# Funktioner
def importAllCases():                   # Läser in dokument och samlar talen
    dfCases = pd.read_csv("Regional_Daily_Cases.csv")
    totCases = dfCases['Sweden_Total_Daily_Cases'].sum()
    
    return totCases

def importDeaths():                     # Läser in dokument och samlar talen
    dfDeaths = pd.read_csv("National_Daily_Deaths.csv")
    totalDeaths = dfDeaths['National_Daily_Deaths'].sum()
    
    return totalDeaths


def Diagrammer(a, b):
    labels = 'Death', 'Alive' 
    sizes = [a, b]
    explode = (0.09, 0)

    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, 
            explode=explode, 
            labels=labels, 
            autopct='%1.1f%%',
            shadow=True, 
            startangle=90)
    ax1.axis('equal')
    
    plt.title("Deaths / Alive ratio ")               #titel för pie charten
    plt.show()  




Diagrammer(importDeaths(), importAllCases())
```
